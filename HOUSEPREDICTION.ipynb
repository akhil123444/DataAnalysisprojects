```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('housing.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
x=df.drop(['median_house_value'],axis=1)
y=df['median_house_value']
```

```python
from sklearn.model_selection import train_test_split

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

```

```python
train_data=x_train.join(y_train)
train_data
```

```python
train_data.hist(figsize=(15,15))
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
train_datal=train_data
train_datal.ocean_proximity=le.fit_transform(train_datal.ocean_proximity)
train_datal
```

```python
train_datal.corr()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_datal.corr(),annot=True)
```

```python
train_datal["total_rooms"]=np.log(train_datal["total_rooms"]+1)
train_datal["total_bedrooms"]=np.log(train_datal["total_bedrooms"]+1)
train_datal["populations"]=np.log(train_datal["population"]+1)
train_datal["housholds"]=np.log(train_datal["households"]+1)
```

```python
train_datal.hist(figsize=(15,8))
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude',y='longitude',data=train_datal,hue='median_house_value',palette='coolwarm')
```

```python
train_datal['bedroom_ratio']=train_datal["total_bedrooms"] / train_datal["total_rooms"]
train_datal['household_rooms']=train_datal["total_rooms"] / train_datal["housholds"]
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_datal.corr(),annot=True)
```

```python
x=train_datal.drop(['median_house_value'],axis=1)
y=train_datal['median_house_value']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
                   
```

```python
from sklearn.ensemble import RandomForestRegressor
forest=RandomForestRegressor()
forest.fit(x_train,y_train)
```

```python
forest.score(x_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
reg.fit(x_train,y_train)
reg.score(x_test,y_test)
```

```python
y_predict=reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
cm=accuracy_score(y_test,y_predict)
cm
```
